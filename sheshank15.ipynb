```python
#include<stdio.h>
int factorial(int a)
{
	int i,f=1;
for(i=a;i>0;i--)
{
	f=f*i;
}
	return f;
	
}
int main()
{
	int x,rem,f=1,n,sum=0;
	scanf("%d",&x);
	n=x;
	while(x!=0)
	{
		rem=x%10;
		f=factorial(rem);
		sum=sum+f;
		x/=10;
	}
	if(sum==n)
		{
			printf("yes");
		}
     else
		{
			printf("no");
		}
		return 0;
	}
		
```
